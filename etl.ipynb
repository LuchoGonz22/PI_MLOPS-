```python
import pandas as pd
```

```python
#Ingestamos los datos
amazon = pd.read_csv("amazon_prime_titles.csv")
netflix = pd.read_csv("netflix_titles.csv")
hulu = pd.read_csv("hulu_titles.csv")
disney = pd.read_csv("disney_plus_titles.csv")
```

# Generamos el campo "id"

```python
amazon.insert(loc=0, column="id", value="a" + amazon["show_id"])
netflix.insert(loc=0, column="id", value="n" + netflix["show_id"])
hulu.insert(loc=0, column="id", value="h" + hulu["show_id"])
disney.insert(loc=0, column="id", value="d" + disney["show_id"])
```

```python
#Verificamos
amazon.head()
```

# Concatenamos los 4 datasets

```python
df = amazon.append(netflix).append(hulu).append(disney)
```

```python
df.isna().sum()
```

```python
df.head()
```

# Reemplazamos los valores nulos del campo rating por “G”

```python
df["rating"].fillna("G", inplace=True)
```

```python
df.isna().sum()
```

# Se cambia el formato de fecha a "AAAA-mm-dd"

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
df.head()
```

# Cambiamos todos los campos que sean texto a minúsculas

```python
df["type"] = df["type"].str.lower()
df["title"] = df["title"].str.lower()
df["director"] = df["director"].str.lower()
df["cast"] = df["cast"].str.lower()
df["country"] = df["country"].str.lower()
df["rating"] = df["rating"].str.lower()
df["duration"] = df["duration"].str.lower()
df["listed_in"] = df["listed_in"].str.lower()
df["description"] = df["description"].str.lower()
```

```python
df.head()
```

# Separamos el campo "duration" en dos

```python
#Separamos la columna
df[["duration_int", "duration_type"]] = df["duration"].str.split(" ", n=1, expand=True)
```

```python
#Eliminamos la columna "duration"
df = df.drop("duration", axis=1)
```

```python
df.head()
```

```python
df.to_csv("Finish_data.csv", index=False)
```

# Ahora trabajamos con los 8 csv

```python
#Ingestamos todos los csv mediante un bucle
ratings = ["1.csv", "2.csv", "3.csv", "4.csv", "5.csv", "6.csv", "7.csv", "8.csv"]
new_df = []
for rating in ratings:
    df1 = pd.read_csv(rating)
    new_df.append(df1)
```

```python
#Unimos los 8 csv en uno solo
r = pd.concat(new_df, axis=0, ignore_index=True)
```

```python
r.shape
```

# Calculamos el "score"

```python
#Usamos "moviesId" y calculamos el promedio
r["avg_r"] = r.groupby(["movieId", "userId"])["rating"].mean().reset_index(drop=True)

# #Creamos la columna "score" e inputamos los valores promediados a la columna "score" mediante "movieId"
r = r.drop_duplicates(subset=['movieId'])
```

```python
r = r.drop(['timestamp'], axis=1)
```

```python
r.head()
```

# Unimos los DataFrame de plataformas con los 8 csv
y los guardamos en formato csv

```python
movies_df = pd.merge(df, r, left_on="id", right_on="movieId", how="inner")
```

```python
movies_df.reset_index().to_csv('movies_dF.csv', index = False)
```
