# Comenzamos visualizando las peliculas

```python
movie_per_year = df["release_year"].value_counts()
movie_per_year
```

```python
sns.histplot(df, x="release_year")
```

# Categoria "Genero"

```python
counts = df["listed_in"].value_counts()
# counts.head()

plt.figure(figsize=(15,5))
plt.pie(counts.values, labels=counts.index)
plt.title("Cantidad de pelicula por genero")
```

# Rating

```python
plt.hist(df.avg_r, log=True)
plt.title("Rating por usuario")
plt.show()
```

```python
print('Rating Promedio =',df.avg_r.mean())
print('Rating maximo =',df.avg_r.max())
print('Rating minimo =',df.avg_r.min())
```
